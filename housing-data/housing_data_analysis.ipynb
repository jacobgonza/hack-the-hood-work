```python
hostle_data = pd.read_csv("Hostel.csv")

lower_rated_hostle = []

for index, row in hostel_data.iterrows():
    if row["rating.band"] == "Superb" or row["rating.band"] == "Fabulous" :
        continue
        
    lower_rated_hostle.append(row)
lower_rated_hostle = pd.DataFrame(lower_rated_hostle)
lower_rated_hostle
```

```python
tokyo_hostel = []

for index, row in hostel_data.iterrows():
    if row["atmosphere"] <= 6:
        tokyo_hostel.append(row)
        
tokyo_hostel = pd.DataFrame(tokyo_hostel)

tokyo_hostel
```
