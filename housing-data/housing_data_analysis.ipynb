```python
hostle_data = pd.read_csv("Hostel.csv")

lower_rated_hostle = []

for index, row in hostel_data.iterrows():
    if row["rating.band"] == "Superb" or row["rating.band"] == "Fabulous" :
        continue
        
    lower_rated_hostle.append(row)
lower_rated_hostle = pd.DataFrame(lower_rated_hostle)
lower_rated_hostle
```

```python
fukuoka_hostel = []

for index, row in hostel_data.iterrows():
    if row["City"] == "Fukuoka-City":
        fukuoka_hostel.append(row)
        
fukuoka_hostel = pd.DataFrame(fukuoka_hostel)

fukuoka_hostel
```

```python
correlation = hostel_data.corr()


fig, ax = plt.subplots(figsize=(20,10))
sns.heatmap(correlation, xticklabels=correlation.columns,yticklabels=correlation.columns, ax = ax)
```

```python
clean_hostel = []

for index, row in hostel_data.iterrows():
    if row["cleanliness"] == 10:
        clean_hostel.append(row)
        
clean_hostel = pd.DataFrame(clean_hostel)

clean_hostel
```

```python
close_hostel = []

for index, row in hostel_data.iterrows():
    if row["Distance"] == "6km from city centre":
        close_hostel.append(row)
        
close_hostel = pd.DataFrame(close_hostel)

close_hostel
```

```python
tokyo_hostel = []

for index, row in hostel_data.iterrows():
    if row["valueformoney"] == 10:
        tokyo_hostel.append(row)
        
tokyo_hostel = pd.DataFrame(tokyo_hostel)

tokyo_hostel
```

```python
tokyo_security_hostel = []

for index, row in hostel_data.iterrows():
    if row["security"] == 10:
        tokyo_security_hostel.append(row)
        
tokyo_security_hostel = pd.DataFrame(tokyo_security_hostel)

tokyo_security_hostel
```

```python
worst_atmosphere_hostel = []

for index, row in hostel_data.iterrows():
    if row["atmosphere"] <= 2:
        worst_atmosphere_hostel.append(row)
        
worst_atmosphere_hostel = pd.DataFrame(worst_atmosphere_hostel)

worst_atmosphere_hostel
```
